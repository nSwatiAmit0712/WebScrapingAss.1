# Q. 1. Problem Statement

To write the python programme to display all the Header tags from www.wilipedia.org

#Importing Necessary Libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/Header'
```

```python
page =requests.get(url)
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content,"html.parser")
print(soup.prettify())
```

```python
scraped_headers = soup.find_all('span',class_= 'mw-headline')
scraped_headers
```

```python
headers = []
for header in scraped_headers:
    header = header.get_text().replace('\n','')
    headers.append(header)
headers
```

```python
data = pd.DataFrame()
data['Header tags'] = headers
data
```

# Q.4 Problem Statement

Write a python programme to display the list of respected former presidents of India

# Import Necessary Libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://presidentofindia.nic.in/former-presidents.htm"
```

```python
page = requests.get(url)
page
```

```python
page.text
```

```python
soup = BeautifulSoup(page.text,"html.parser")
print(soup.prettify())
```

```python
scraped_list = soup.find_all('div',class_='presidentListing')
scraped_list
```

```python
names =[]
for name in scraped_list:
    name = name.get_text().replace('\n',"")
    names.append(name)
names
```

```python
data = (zip(names))
df = pd.DataFrame(data,columns=['Name'])
df
```
